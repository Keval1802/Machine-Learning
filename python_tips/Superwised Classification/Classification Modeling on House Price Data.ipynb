# Prepare data for machine learning

```python
# separate the target variable and independent variable.
target = df_houseprice['Sold']
```

```python
#independent variable
df_independent = df_houseprice.drop('Sold',axis = 1)
df_independent
```

 ## seprate the categorical column from independent variable


```python
df_cat = df_independent.select_dtypes(include = "object")
df_cat
```

```python
#encode the categorical variable
encoded = pd.get_dummies(df_cat)
```

```python
encoded
```

## Scale the numerical data

```python
df_num = df_independent.select_dtypes(include = 'number')
```

```python
df_num.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
scale = ss.fit_transform(df_num)
scale
```

```python
df_scale = pd.DataFrame(scale,columns = df_num.columns)
df_scale = df_scale.reset_index()
df_scale
```

```python
X = pd.concat([df_num,encoded],axis = 1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, target, random_state = 10 , test_size = 0.2)
X_train.shape
y_train.shape

X_test.shape
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

tree_model = DecisionTreeClassifier()

tree_model.fit(X_train,y_train)

prediction = tree_model.predict(X_test)

accurecy = accuracy_score(y_test,prediction)
accurecy
```
